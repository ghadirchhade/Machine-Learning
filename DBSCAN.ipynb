```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
n_samples = 500
radius = 1.0
theta = np.linspace(0, 2*np.pi, n_samples)

# Générer les coordonnées polaires
x = radius * np.cos(theta)
y = radius * np.sin(theta)

# Ajouter du bruit (si nécessaire)
noise_factor = 0.1
x += np.random.normal(0, noise_factor, n_samples)
y += np.random.normal(0, noise_factor, n_samples)

# Afficher le jeu de données circulaire
plt.scatter(x, y)
plt.title('Jeu de données circulaire')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"x":x,"y":y})
df
```

```python
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=2)
Kmeans.fit(df)
```

```python
plt.scatter(df['x'], df['y'], c=Kmeans.labels_, s=50, cmap='viridis')
centers = Kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.3,min_samples=10)
dbscan.fit(df)
```
