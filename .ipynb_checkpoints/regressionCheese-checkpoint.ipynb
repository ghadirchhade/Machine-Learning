```python
import pandas as pd
import seaborn as sns
```

```python
fromage=pd.read_table("../datasets/fromage.txt",index_col=0)
fromage.head()
```

```python
sns.set(rc={'figure.figsize':(10,4)})
df_corr = fromage.corr()
ax = sns.heatmap(df_corr, xticklabels = fromage.columns , 
                 yticklabels = df_corr.columns, annot=True, cmap = 'coolwarm')
```

```python
X=fromage.drop(["calories"],axis=1)
X.head()
```

```python
y=fromage.calories
y.head()
```

# NORMALIZATION

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_CR=pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns)
X_train_CR.head()
```

```python
X_train_CR.shape
```

# @ MULTIPLE LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X_train_CR,y_train)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

# EQUATION MULTIPLE LINEAR 

```python
coef=pd.DataFrame(lr_model.coef_,columns=["COEFICCIENT"],index=X_train.columns)
coef.loc["constante"]=lr_model.intercept_
coef
```

```python
X_test_CR=pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)
X_test_CR.head()
```

```python
y_pred=lr_model.predict(X_test_CR)
y_pred,X_test.shape
```

```python
scaler.var_,scaler.mean_
```

# METRICS

```python
from sklearn.metrics import mean_squared_error,r2_score
print("RMSE:",mean_squared_error(y_test,y_pred,squared=False))
print("R square:",r2_score(y_test,y_pred))
```

### --->as we see in the previous cell R square is close from 1 --> good model

# @REGRESSION RIDGE

### 1)without GridSearchCV

```python
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=0)
ridge_model.fit(X_train_CR,y_train)
```

```python
y_pred=ridge_model.predict(X_test_CR)
y_pred
```

```python
print("RMSE:",mean_squared_error(y_test,y_pred,squared=False))
print("R square:",r2_score(y_test,y_pred))
#like the multiple regression because alpha=0-->it dones the optimal parameters because the penalty=0
```

# EQUATION RIDGE WITHUT GRIDSEARCHCV

```python
coef=pd.DataFrame(ridge_model.coef_,columns=["coefficient"],index=X_train.columns)
coef.loc["constante"]=ridge_model.intercept_
coef
```

### 2)with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV,ShuffleSplit
import numpy as np
parameters={"alpha":np.arange(start=0,stop=10,step=0.05)}
ridge_model=Ridge()
ridge_model=GridSearchCV(ridge_model,parameters,scoring="r2",verbose=1)
ridge_model.fit(X_train_CR,y_train)
```

```python
ridge_model.best_params_,ridge_model.best_score_
```

```python
y_pred=ridge_model.predict(X_test_CR)
print("RMSE:",mean_squared_error(y_test,y_pred,squared=False))
print("R square:",r2_score(y_test,y_pred))
```

```python
ridge_model.best_estimator_
```

# EQUATION RIDGE WITH GRIDSEARCHCV

```python
coef=pd.DataFrame(ridge_model.best_estimator_.coef_,columns=["coefficient"],index=X_train.columns)
coef.loc["constante"]=ridge_model.best_estimator_.intercept_
coef
```

# LASSO EQUATION WITH GRIDSEARCHCV

```python
coef=pd.DataFrame(lasso_model.best_estimator_.coef_,
                  index=X_train.columns,
                  columns=["coefficient"])
coef.loc["constante"]=lasso_model.best_estimator_.intercept_
coef
```

# @ELASTICNET REGRESSION

### 1)without GridSearchCV:

```python
from sklearn.linear_model import ElasticNet
elastic_model=ElasticNet(alpha=0.5,l1_ratio=0.8)
elastic_model=elastic_model.fit(X_train_CR,y_train)
y_pred=elastic_model.predict(X_test_CR)
#l1_ratio close from 1-->use lasso
print("RMSE:",mean_squared_error(y_test,y_pred,squared=False))
print("R square:",r2_score(y_test,y_pred))
```

```python
coef=pd.DataFrame(elastic_model.coef_,
                  index=X_train.columns,
                  columns=["coefficient"])
coef.loc["coefficient"]=elastic_model.intercept_
coef
```

### 2)with GridSearchCV

```python
parameters={"alpha":np.arange(0,10,1),
            "l1_ratio":np.arange(0,10,1)}
elastic_model=ElasticNet()
elastic_model=GridSearchCV(elastic_model,parameters,scoring="r2",verbose=1,cv=2)
elastic_model.fit(X_train_CR,y_train)
```

```python
y_pred=elastic_model.predict(X_test_CR)
print("RMSE:",mean_squared_error(y_test,y_pred,squared=False))
print("R square:",r2_score(y_test,y_pred))
```

```python
elastic_model.best_params_,elastic_model.best_score_
```

```python
elastic_model.best_estimator_
```

# ELASTICNET EQUATION

```python
coef = pd.DataFrame(elastic_model.best_estimator_.coef_ ,
                    index = X_train.columns, columns=['Coefficient'])
coef.loc['Constante'] = elastic_model.best_estimator_.intercept_
coef
```

```python
from sklearn.linear_model import ElasticNet
elastic_lasso = ElasticNet(alpha=5, l1_ratio=1)
elastic_lasso = elastic_lasso.fit(X_train_CR,y_train)
y_pred = elastic_lasso.predict(X_test_CR)
print("RMSE : " ,mean_squared_error(y_test, y_pred, squared= False))
print("R square : ",r2_score(y_test, y_pred))
```

```python
coef = pd.DataFrame(elastic_lasso.coef_ ,
                    index = X_train.columns, columns=['Coefficient'])
coef.loc['Constante'] = elastic_lasso.intercept_
coef
```
