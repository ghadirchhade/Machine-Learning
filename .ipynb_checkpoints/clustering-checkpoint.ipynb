# Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns; sns.set()  # for plot styling
from matplotlib.patches import Ellipse
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score
```

# K_MEANS

### we take only 2 features from the df without label to be able to visualisation the result (2 dimension)

```python
X=df_fromage[["calories","folates"]]
X.head()
```

```python
X.plot(kind="scatter",x="calories",y="folates",c="b")
#plt.scatter(X["calories"],X["folates"])
```

```python
X.shape
```

### -->here we want to see the result of the clustering without normalisation of features

```python
kmean=KMeans(n_clusters=3,random_state=0)#we take 3 by default
kmean.fit(X)#we apply fit on the features df,mafi train w test hon
```

```python
kmean.labels_
#display the classes
#we remark 3 classes(0,1,2) like the nb of clusters
```

```python
kmean.cluster_centers_
#display the coordinates(calories,folates) of the last centers of each cluster
```

```python
plt.scatter(X["calories"],X["folates"],c=kmean.labels_,s=50,cmap="viridis")
#c means that we want to colore the scatter selon the labels(Each cluster will be represented by a different color on the plot)
#cmap indicate the coloring map( that goes from yellow to green to blue, with increasing values)
#s=50-->This parameter specifies the size of the points in the scatter plot. In this case, all points will have a size of 50.
centers=kmean.cluster_centers_
#we add to the scatter a new scatter(centers) above it
plt.scatter(centers[:,0],centers[:,1],c="black",s=200,alpha=0.5)
#centers[:,0]-->the first column from the centers array it will be placed on the x-axis
```

### as we see it takes the centers out from our dataset

### note:without normalisation we see that the result is soo bad and it takes the clusters vertically-->its not a good clustering
### to obtain a good clustering we must increase the inter-cluters and decrease the intra-clusters and tht its not applyed here

```python
#add the labels to the df as a new column
#we add it only for compare at the end when we normalize the features
clusters=kmean.labels_
df_fromage["kmeans_1"]=clusters
df_fromage.head()
```

```python
df_fromage.drop(["kmean_1"],axis=1,inplace=True)
```

### -->here we want to see the result of the clustering with normalisation of features

```python
scaler=StandardScaler()
XCR=scaler.fit_transform(X)
XCR=pd.DataFrame(XCR,columns=X.columns)
XCR.head()
```

```python
result=KMeans(n_clusters=3,random_state=0)
result.fit(XCR)
```

```python
result.labels_
```

```python
result.cluster_centers_
```

```python
plt.figure()
plt.scatter(XCR["calories"],XCR["folates"],c=result.labels_,cmap="viridis")
centers=result.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c=np.unique(result.labels_),lw=2,edgecolors ="k",cmap='viridis',s=200, alpha=0.5);
```

### as we see the scatter above with normalised features it returns a good clustering:minimised intra-clusters and maximised inter-clusters not like without normalisation

```python
df_fromage["KMeans_1CR"]=result.labels_
df_fromage
```

```python
#we want to see the difference before and after normalisation using crosstab
pd.crosstab(df_fromage.kmeans_1,df_fromage.KMeans_1CR,colnames=["KMeans"],rownames=["KMeans_CR"],margins=True)
```

# !!!!!!!!!NOW:THE IMPORTANT PART:
## CHOICE THE NUMBER OF CLUSTERS K:

```python
from matplotlib.patches import Ellipse
K = np.arange(2,10,1)
Sum_of_squared_distances = []
for k in K:
    km_char = KMeans(n_clusters=k)
    km_char = km_char.fit(XCR)
    Sum_of_squared_distances.append(km_char.inertia_)
    
print(Sum_of_squared_distances)

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.xticks(np.arange(2, 10, step=1))
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')

#On affiche l'ellipse 
ax = plt.gca()

ellipse = Ellipse(xy=(3, 9), width=1, height=5, 
                        edgecolor='r', fc='None', lw=2)
ax.add_patch(ellipse)
```

# CLASSIFICATION ASCENDANTE HIERARCHIQUE

### 1)using scipy library to implement the dendogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
Z = linkage(XCR,method='ward',metric='euclidean')
#merging clusters based on the ward method(minimize the variance)
plt.title("CAH")
dendrogram(Z,labels= X.index.tolist(),orientation='top',color_threshold=2)
#index=cheese's name
plt.show()
```

```python
Z = linkage(XCR,method='ward',metric='euclidean')
Z,Z.shape
```

```python
plt.title("CAH with 3 clusters")
threshold=4.5
dendrogram(Z,labels= X.index.tolist(),orientation='top',color_threshold=threshold)
ax=plt.gca()
ax.axvline(x=threshold, ymin=0, ymax=1,ls="--",color="r")
plt.text(threshold,0,str(threshold),rotation=0,color="r")
plt.show()
```

```python
groupes_cah = fcluster(Z ,t=7,criterion='distance')
#groupes_cah = fcluster(Z ,t=4,criterion='maxclust')
print(groupes_cah)

df_fromage['CAH'] = groupes_cah
df_fromage.head(20)
```

### 2)using scikit learn library

```python
X=df_fromage.iloc[:,0:9]
X.head()
```

```python
scaler=StandardScaler()
fromage_CR=scaler.fit_transform(X)
```

```python
threshold = 8
Z = linkage(fromage_CR,method='ward',metric='euclidean')
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(1, 1, 1)
plt.title('CAH avec seuil  6')
dendrogram(Z,labels= X.index.tolist(),orientation='top',color_threshold=threshold)
plt.axvline(x=threshold,linestyle = '--')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
groupes_cah = fcluster(Z ,t=8,criterion='distance')
groupes_cah = groupes_cah-1
print(groupes_cah)
df_fromage['CAH'] = groupes_cah

silhouette_score(X,groupes_cah)
#print(len(groupes_cah),np.shape(fromage_CR))
```

```python
agg = AgglomerativeClustering(n_clusters=3, linkage="ward")
agg.fit(X)
agg.labels_
```

```python
plt.figure()
plt.scatter(X['calories'], X['folates'], c=groupes_cah, s=50, cmap='viridis')
plt.xlabel("Calories")
plt.ylabel("Folates");
```

```python
df_fromage["CAH_CR"]=agg.labels_
df_fromage.head(20)
```
